```python
from deepface import DeepFace
import matplotlib.pyplot as plt
import cv2
```

```python
backends= ["opencv" , "ssd" , "retinaface" , ]

face = DeepFace.detectFace(
    "Z:\\\\face recognized project work\\\\ronaldo\\\\ronaldo2.jpeg", 
     detector_backend="opencv"
)
```

```python
face = face.squeeze(0)

# Display the image
plt.imshow(face)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, axs = plt.subplots(2,2, figsize=(15, 10))
axs = axs.flatten()

for i, b in enumerate(backends):
    try:
        axs[i].set_title(b)
        img_path = r"Z:\\\\face recognized project work\\\\ronaldo\\\\ronaldo2.jpeg"
        face = DeepFace.extract_faces(img_path, detector_backend=b)
        print(f"Face detection output for {b}: {face}")
        if face is not None:
            for face_data in face:
                face_array = np.array(face_data['face'])
                print(f"Face shape: {face_array.shape}")
                axs[i].imshow(face_array, cmap='gray')
                axs[i].set_title(b)
                axs[i].axis("off")
        else:
            print(f"No face detected for {b}")
    except Exception as e:
        print(f"Error: {e}")

plt.show()
```
